```
import requests
import urllib.request
import time
import os
from bs4 import BeautifulSoup
```

```
url = 'https://en.wikipedia.org/wiki/2020_coronavirus_pandemic_in_Spain'
response = requests.get(url)
website_url = response.text
```

```
soup = BeautifulSoup(website_url,'lxml')
print(soup.prettify())
```

```
#my_table = soup.find('table: ', {'class': 'wikitable sortable'})
#my_table
```

```
tables = soup.find_all('table',class_='mw-collapsible')
```

```
headingsinterest = []
```

```
for table in tables:
    ths = table.find_all('th')
    headings = [th.text.strip() for th in ths]
    print(headings)
    if headings[0] == 'Reported by':
        headingsinterest = headings
        break
```

```
rows = table.findAll('tr')
row_lengths = [len(r.findAll(['th','td'])) for r in rows]
ncols = max(row_lengths)
nrows = len(rows)
data = []
for i in range(nrows):
    rowD = []
    for j in range(ncols):
        rowD.append('')
    data.append(rowD)
for i in range(len(rows)):
        row = rows[i]
        rowD = []
        cells = row.findAll(["td", "th"])
        for j in range(len(cells)):
            cell = cells[j]

            cspan = int(cell.get('colspan', 1))
            rspan = int(cell.get('rowspan', 1))
            l = 0
            for k in range(rspan):
                try:
                    while data[i + k][j + l]:
                        l += 1
                except:
                    print(i,k,j,l)
                for m in range(cspan):
                    cell_n = j + l + m
                    row_n = i + k
                    cell_n = min(cell_n, len(data[row_n])-1)
                    data[row_n][cell_n] += cell.text
                    #print(cell.text)

        data.append(rowD)
```
